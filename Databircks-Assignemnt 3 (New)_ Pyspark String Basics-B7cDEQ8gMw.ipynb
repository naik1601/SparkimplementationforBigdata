## Basic Level Questions

1. Convert the 'Region' column to lowercase.

```python
from pyspark.sql.functions import initcap, lower, upper, col

df1 = df.withColumn("Region", lower(col("Region")))
display(df1)
```

2. Convert the 'Country' column to uppercase.

```python
# Your code here
from pyspark.sql.functions import initcap, lower, upper, col
df2=df.withColumn('Country',upper(col('Country')))
df2.display()
```

3. Apply initcap to the 'ItemType' column.

```python
# Your code here
from pyspark.sql.functions import initcap,upper,lower,col
df3=df.withColumn('Itemtype',initcap(col('Itemtype')))
df3.display()
```

4. Concatenate 'Region' and 'Country' columns with a hyphen between them.

```python
# Your code here
from pyspark.sql.functions import concat,col,concat_ws
df4=df.select(concat_ws('-',col('Region'),col('Country')))
df4.display()
```

5. Find the length of the 'OrderID' column.

6. Trim any leading or trailing spaces from the 'ItemType' column.

```python
# Your code here
from pyspark.sql.functions import length, col,trim

df5 = df.select(trim(col("ItemType")).alias("ItemType"))
display(df5)
```

7. Left pad the 'OrderID' with zeros to ensure it's always 10 characters long.

```python
# Your code here
from pyspark.sql.functions import col, lpad

df2 = df.withColumn("OrderID", lpad(col("OrderID").cast("string"), 10, "0"))
display(df2)
```

8. Extract the first 3 characters from the 'Country' column.

```python
# Your code here
from pyspark.sql.functions import col,substring
df8=df.withColumn('Country',substring(col('Country'),0,3))
df8.display()
```

9. Concatenate 'ItemType' and 'OrderPriority' with a comma and space between them.

```python
# Your code here
from pyspark.sql.functions import col,concat_ws
df9=df.select(concat_ws(', ',col('ItemType'),col('OrderPriority')))
df9.display()
```

10. Replace any occurrence of 'and' with '&' in the 'Region' column.

```python
from pyspark.sql.functions import col, regexp_replace

df2 = df.withColumn(
    "Region",
    regexp_replace(col("Region"), r"(?i)\band\b", "&")
)

display(df2)
#(?i) = case-insensitive (and, AND, And)
#\band\b = only the word “and” (won’t change parts of other words)


```

11. Create a new column 'OrderCategory' that shows 'High' for High priority, 'Medium' for Medium, and 'Low' for Low priority orders.

```python
from pyspark.sql.functions import col, when

df11 = df.withColumn(
    "OrderCategory",
    when(col("OrderPriority") == "H", "High")
    .when(col("OrderPriority") == "M", "Medium")
    .when(col("OrderPriority") == "L", "Low")
)

display(df11)
```
