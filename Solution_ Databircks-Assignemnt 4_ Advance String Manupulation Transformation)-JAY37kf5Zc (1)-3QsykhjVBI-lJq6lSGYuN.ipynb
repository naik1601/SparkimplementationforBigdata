<b> What is the abbreviated version of each employee's last name, with the first letter and the two following letters included?

```python
from pyspark.sql.functions import *
df1=spark.read.csv("/FileStore/EmployeeData_fWRqkJfIxK.csv",header=True,inferSchema=True)
display(df1)
df2=df1.withColumn("first_and_char",df1.last_name.substr(1,3))
display(df2)
```

```python
df100=df1.createOrReplaceTempView("Employee")
df101=spark.sql("Select *, case when last_name is not null then left(last_name,3) else last_name END as first_and_char from Employee")
display(df101)
```

<b> What is the domain name of each employee's email address, so that their emails can be filtered and routed to the correct department?

```python
df3=df1.select(split('email','@').getItem(1).alias("email_domain"),"first_name").select("email_domain","first_name").show()
```

```python
df102=spark.sql("select substring(email,CHARINDEX( '@', email) +1,LEN(email)) as email_domain,first_name from Employee")
display(df102)
```

<b> In an effort to protect their privacy, what is each employee's name spelled backwards?

```python
df4=df1.withColumn('backwards_name',reverse(concat_ws(' ',df1.first_name,df1.last_name))).show()
```

```python
df103=spark.sql("select *,REVERSE(CONCAT(first_name,' ',last_name)) as backwards_name from Employee")
display(df103)
```

<b> In order to make their email addresses harder to guess, what is each employee's email address spelled backwards?

```python
df5=df1.withColumn("backwards_email",reverse(concat_ws(' ',df1.email))).show()
```

```python
df104=spark.sql("select *,REVERSE(email) as backwards_email from Employee")
display(df104)
```

<b> What is the country code of each employee's email domain, which can help identify global market opportunities?

```python
# Write your code using the Spark Function
df8=df1.select(expr("substring((email),length((email)) - length(substring_index((email), '\.', -1)),length((email)))").alias('countrycode'))
display(df8)
```

```python
df106=spark.sql("select email,substring(email,CHARINDEX( '\.', email)+0,LEN(email)) as countrycode,first_name from Employee")
display(df106)
```

<b> What is the position of the first occurrence of a dot ('.') character in each employee's email address, which can be used to standardize email addresses in a database?

```python
# Write your code using the Spark Function
df9=df1.withColumn("first_dotposition_in_email",instr(df1.email, '.')).select("email","first_dotposition_in_email")
display(df9)
```

```python
df107=spark.sql("select email, CHARINDEX( '.', email) as first_dotposition_in_email from Employee")
display(df107)
```

```python
df7=df1.fillna("",subset="first_name").withColumn("first_name_upper",initcap("first_name")).select("first_name_upper").show()
```
