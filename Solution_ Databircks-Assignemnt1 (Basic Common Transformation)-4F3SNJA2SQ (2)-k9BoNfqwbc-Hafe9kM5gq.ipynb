<b>How many employees are in the input file?

```python
from pyspark.sql.functions import *
df1=spark.read.csv("/FileStore/EmployeeData_8xfZvOlzqL-1.csv",header=True,inferSchema=True)
display(df1)
df1.count()
```

```python
df1.createOrReplaceTempView("EmployeeData")
df100=spark.sql("Select count(*) from EmployeeData")
display(df100)
```

<b> What is the average salary of all the employees?

```python
df2=df1.agg(avg("salary")).show()
```

```python
df101=spark.sql("Select AVG(Salary) from EmployeeData")
display(df101)
```

<b> How many male and female employees are there in the input file?

```python
df3=df1.filter((df1.gender=='Male') | (df1.gender=='Female')).groupBy(df1.gender).count().show()
```

```python
df102=spark.sql("Select gender, count(gender) as count from EmployeeData where gender not in ('null') group by gender")
display(df102)
```

<b> Who is the employee with the highest salary?

```python
df4=df1.agg(max("salary")).first()[0]
df5=df1.filter((df1.salary==df4)).select("first_name","last_name","salary").show()
```

```python
df103=spark.sql("Select first_name,last_name,salary from EmployeeData order by salary Desc Limit 1 ")
display(df103)
```

<b> Who are the top 5 employees with the highest salaries?

```python
from pyspark.sql.window import Window
w=Window().orderBy(df1.salary.desc_nulls_last())
df8=df1.withColumn("row_num",row_number().over(w))
df9=df8.filter(df8.row_num<=5).show()
```
