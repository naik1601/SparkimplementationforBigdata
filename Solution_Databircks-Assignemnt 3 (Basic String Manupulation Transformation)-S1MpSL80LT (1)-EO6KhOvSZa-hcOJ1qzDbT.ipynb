<b> How many employees have a first name that starts with the letter 'J'?

```python
from pyspark.sql.functions import *
df1=spark.read.csv("/FileStore/EmployeeData_HhTLiz9LdC.csv",header=True,inferSchema=True)
display(df1)
df2=df1.filter(df1.first_name.startswith('J'))
df2.count()
```

```python
df100=df1.createOrReplaceTempView("Employee")
df101=spark.sql("Select * from Employee where first_name like 'J%'")
display(df101)
```

<b> How many employees have a last name that ends with the letter 's'?

```python
df3=df1.filter(df1.last_name.endswith('s'))
df3.count()
```

```python
df102=spark.sql("Select * from Employee where last_name like '%s'")
display(df102)
```

<b> What is the length of the longest first name?

```python
df4=df1.withColumn("len_longest_fist_name",length(df1.first_name)).groupBy().max("len_longest_fist_name")
display(df4)
```

```python
df103=spark.sql("Select len(first_name) as len_longest_fist_name from Employee order by len_longest_fist_name DESC limit 1")
display(df103)
```

<b> What is the length of the shortest last name?

```python
df5=df1.withColumn("len_shortest_last_name",length(df1.last_name)).groupBy().min("len_shortest_last_name")
display(df5)
```

```python
df104=spark.sql("Select len(last_name) as len_shortest_last_name from Employee where last_name is not NULL order by len_shortest_last_name  limit 1")
display(df104)
```

<b>What is the email domain of the employee with the highest salary?

```python
highest_salary_email_dom=df1.select(split("email","@").getItem(1).alias("email_domain"),"salary").orderBy(desc("salary")).limit(1).select("email_domain").collect()[0][0]
display(highest_salary_email_dom)
```

```python
df110=spark.sql("select substring(email,CHARINDEX( '@', email) +1,LEN(email)) as email_domain from Employee where salary not in ('null') order by salary desc limit 1 ")
display(df110)
```
