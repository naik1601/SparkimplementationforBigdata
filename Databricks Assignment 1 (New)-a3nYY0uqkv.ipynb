## Question 1
Select only the 'OrderID', 'Region', and 'TotalRevenue' columns from the dataset.

## Question 2
Find all orders from the 'Europe' region with a total revenue greater than $500,000.

```python
from pyspark.sql.functions import col

df6 = df.filter((col("TotalRevenue").cast("double") > 500000) & (col("Region") == "Europe"))
display(df6)
```

## Question 3
List the distinct item types sold in the 'Middle East and North Africa' region.

```python
from pyspark.sql.functions import col


df2 = (
    df.filter(col("Region") == "Middle East and North Africa")
      .select("ItemType")
      .distinct()
)

display(df2)

```

## Question 4
Select orders that are either 'High' priority or have a total profit greater than $100,000.

```python
from pyspark.sql.functions import col

df3 = df.filter((col("OrderPriority") == "H") & (col("TotalProfit").cast('double') > 100000))
display(df3)
```

## Question 5
Create a new column called 'RevenueCategory' that categorizes total revenue as 'Low' if less than $100,000, 'Medium' if between $100,000 and $500,000, and 'High' if greater than $500,000.

```python
# from pysarpk.sql.functions import col
from pyspark.sql.functions import col, when

df5 = df.withColumn(
    'RevenueCategory',
    when(col('TotalRevenue').cast('double') > 500000, 'High')
    .when((col('TotalRevenue').cast('double') >= 100000) & (col('TotalRevenue').cast('double') <= 500000), 'Medium')
    .otherwise('Low')
)

display(df5)
```

## Question 6
Remove the 'UnitCost' column from the dataset and rename 'TotalRevenue' to 'Revenue'.

```python
from pyspark.sql.functions import col

df6 = df.drop("UnitCost")
df6.show()
df7 = df6.withColumn("Revenue", col("TotalRevenue"))
df7.show()
```

## Question 8
Create a new column 'ProfitMargin' that calculates the percentage of profit relative to revenue (TotalProfit / TotalRevenue * 100).

## Question 10
Find all orders where the 'UnitPrice' is greater than the average 'UnitPrice' of all products.

```python
# Your code here

from pyspark.sql.functions import col, avg
avg_price=df.select(avg('UnitPrice').alias('avg_price')).first()["avg_price"]
df10=df.filter(col('Unitprice')>avg_price)
display(df10)
```
